**¿Qué es Kernel?**

Un kernel es una función utilizada en SVM para ayudar a resolver problemas. Proporcionan atajos para evitar cálculos complejos. Lo sorprendente del kernel es que podemos ir a dimensiones más altas y realizar cálculos suaves con la ayuda de él.

Podemos subir a un número infinito de dimensiones usando kernels.

A veces, no podemos tener un hiperplano para ciertos problemas. Este problema surge cuando subimos a dimensiones superiores y tratamos de formar un hiperplano.

Un núcleo ayuda a formar el hiperplano en la dimensión superior sin aumentar la complejidad.

**Tipos de kernel**

image.png

1. Función de kernel polinómico
El núcleo polinómico es una representación general de núcleos con un grado de más de uno. Es útil para el procesamiento de imágenes.

Hay dos tipos de esto:

Función de kernel polinómico homogéneo
K(xi,x j) = (xi.x j)d, donde '.' es el producto de puntos de ambos números y d es el grado del polinomio.

Función de kernel polinómico no homogéneo
K(xi,x j) = (xi.x j + c)d donde c es una constante.

2. Función gaussiana del kernel RBF
RBF es la función de base radial. Esto se utiliza cuando no hay conocimiento previo sobre los datos.

Se representa como:

K(xi,x j) = exp(-γ|| xi – xj||) número arábigo

3. Función del núcleo sigmoide
Esto se utiliza principalmente en redes neuronales. Podemos mostrar esto como:

K(xi,x j) = tanh(αxay + c)

4. Función de kernel tangente hiperbólica
Esto también se utiliza en redes neuronales. La ecuación es:

K(xi,x j) = tanh(kxi.x j + c)

5. Función lineal del núcleo
Este kernel es unidimensional y es la forma más básica de kernel en SVM. La ecuación es:

K(xi,x j) = xi.x j + c

6. Función graph Kernel
Este núcleo se utiliza para calcular el interior en gráficos. Miden la similitud entre pares de gráficos. Contribuyen en áreas como bioinformática, quimioinformática, etc.

7. Función de kernel de cadena
Este kernel funciona sobre la base de cadenas. Se utiliza principalmente en áreas como la clasificación de textos. Son muy útiles en minería de textos, análisis genómico, etc.

8. Función de kernel de árbol
Este núcleo está más asociado con la estructura del árbol. El kernel ayuda a dividir los datos en formato de árbol y ayuda al SVM a distinguir entre ellos. Esto es útil en la clasificación del lenguaje y se usa en áreas como la PNL.

```python
from sklearn import datasets
data_iris=datasets.load_iris()
```

```python
data_iris.data.shape
```

```python
data_iris.data
```

```python
data_iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_iris.data, data_iris.target, test_size=0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
NVG=GaussianNB()
```

```python
NVG.fit(X_train,y_train)
```

```python
valores_pred=NVG.predict(X_test)
valores_pred
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,valores_pred)
```

```python
metrics.confusion_matrix(y_test,valores_pred)
```

```python
from sklearn.linear_model import LogisticRegression
modelo_reglog = LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
modelo_entrenado = modelo_reglog.fit(X_train, y_train)
```

```python
y_pred = modelo_entrenado.predict(X_test)
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```
