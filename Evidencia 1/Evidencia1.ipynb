```python
b = np.vstack([2,5])
x = np.vstack([-1,4])
```

```python
lambd = np.dot(x.T,b)/np.linalg.norm(b)**2 
print(lambd)
```

```python
vector_proyectado = b*lambd
print(vector_proyectado)
```

```python
matriz_proyeccion = np.dot(b,b.T)/np.linalg.norm(b)**2 #bb^T/||b||^2
print("Matriz de proyection\n",matriz_proyeccion)
```

# Punto 3
image.png

```python
raicesAndAutovectores = np.linalg.eig(matriz_proyeccion)
print(raicesAndAutovectores)
```

```python
autovector_1= np.linalg.eig(matriz_proyeccion)[1][:,1]
autovector_2= np.linalg.eig(matriz_proyeccion)[1][:,0]
print(autovector_1)
print(autovector_2)
```

```python
import matplotlib.pyplot as plt 
# Figure setup.
fig = plt.figure()
ax = fig.add_subplot(111)
scale = 25
origin2D = np.vstack([0,0])
plt.axis([-scale/2,scale/2,-scale/2,scale/2])
#plt.title("Vectores Original  y autovectores generados.")
plt.quiver(*origin2D,*matriz_proyeccion[:,0],scale=scale, label='Vectores Originales')
plt.quiver(*origin2D,*matriz_proyeccion[:,1],scale=scale)
# Figure setup.


# Eigenvectors.

plt.quiver(*origin2D, *(autovector_1*8), scale=scale,color="red", label='Autovectores_1')
plt.quiver(*origin2D, *(autovector_2*4), scale=scale,color="blue", label='Autovectores_2')


ax.set_aspect('equal', adjustable='box')
plt.legend();
```

```python
x, y = sym.symbols("x y")
funcion= (x**2+7*y-8) + (9*x+y**2)**2
funcion
```

```python
#gereramos un conjunto de datos a partir de la funcion
l = 4
n = 150
linspace = np.linspace(-l,l,n) 

#
X,Y = np.meshgrid(linspace,linspace)

#
Z = f(X,Y)
```

```python
f1yLam = sym.lambdify('x,y',f1y)
f1 = [f1x,f1y]
gradiente = sym.lambdify('x,y',f1)
def gradientDescent(gradiente, punto_inicio, tamao_paso = .025, precision = .01, max_iter = 3000, initError = 10):
    # configuracion inicial
    x,y = punto_inicio.astype(float)
    currIter, iterCoords = 0, [np.hstack([x,y])]
    error = initError
    currX = np.array([x,y])

    # inicio de las iteraciones
    
    while np.linalg.norm(error) > precision and currIter<max_iter:
        currIter+=1 # se incrementa el contador
        prevX = currX.copy() # registra el ultimo valor de error
        
        # Gradient descent equation
        currX -= tamao_paso * np.array(gradiente(x,y)) # New X = (last) minus (step size * function gradient(last)).
        x,y = currX[0].copy(), currX[1].copy() # Update X.
        # Gradient descent equation
        
        error = currX - prevX # Error based on how much movement new X made based on gamma and gradient.
        iterCoords.append(np.hstack([x,y])) # Track X for plotting.
        
    return currIter, currX, np.vstack(iterCoords)


```
