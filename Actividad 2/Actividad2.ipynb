```python
import numpy as np
import matplotlib.pyplot as plt
```

###Genera dos vectores en R2 y represéntalos en un plano utilizando la librería Matplotlib.



```python
#Vectores en R2
vector1 = np.array([5,3])
vector2 = np.array([7,8])
print(vector1)
print(vector2)
```

```python
# Funcion para graficar vectores 
def graficarVectores(vecs, cols, alpha = 1):
  #Crea un lienzo vacio 
  plt.figure()
  #Inidica el color de los ejes  de X and Y 
  plt.axvline(x = 0, color = "blue")
  plt.axhline(y = 0, color = "grey")

  for i in range(len(vecs)):
        #regrensa la longitud de la lista len(vecs) -> ejemplo 2
        #Obtiene un hacerlo iterable range(len(vecs) -> ejemplo rango del 0 al 2 solo haces dos bucles i= 1 and i=2
        x = np.concatenate([[0,0], vecs[i]])
        #Indicamos el origen del vector en este caso 0,0 y le pasamos el vector
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                  angles='xy', scale_units='xy', scale=1,
                   color=cols[i],
                  alpha = alpha)
```

```python
graficarVectores([vector1,vector2], ['blue', 'red'])
#Indica los limites 
plt.xlim(-1, 10)
plt.ylim(-1, 10)
```

###Calcula el producto punto entre los vectores generados.

```python
#Producto punto entre vectores
producto_punto = np.dot(vector1,vector2)
print("Producto punto es ",producto_punto)

producto_interno = np.inner(vector1,vector2)
print("Producto interno es ",producto_interno)
#El producto punto o producto interno es el mismo en vectores la funcion dot o inner regresa el mismo resultado pero esto 
#cambia cuando cuando trabajamos con matrices 
```

```python
#Ahora convertiremos nuestros vectores a forma matricial 
#Nos crear un vector columna matricial
vectorToMatricial1 = np.vstack(vector1)
vectorToMatricial2 = np.vstack(vector2)
print(vectorToMatricial1)
print(vectorToMatricial2)
```

```python
#Ahora haremos el producto punto 
try:
 productoPunto2 = np.dot(vectorToMatricial1,vectorToMatricial2)
except:
  print("Aqui nos lanzara un expecion por que estamos haciendo un producto punto de 2x1 por 2x1 y esto rompe la regla")
```

```python
#Ahora la form de solucionarlo es el primero vector por la tranpuesta del segundo vector 
productoPunto2 = np.dot(vectorToMatricial1,vectorToMatricial2.T)
print("Producto punto de los vectores matriciales\n",productoPunto2)
#Otra forma de hacerlo es utlizar la funcion iner
productoPunto2 = np.inner(vectorToMatricial1,vectorToMatricial2)
print("Producto punto de los vectores matriciales utilizando la funcion inner\n",productoPunto2)
```

###Encuentra el ángulo que forman los vectores entre sí y genera un nuevo vector que sea ortogonal a cualquiera de los dos primeros.

```python
#Los vectores son ortogonales siempre y cuando su producto punto de los vectores sea cero
#La una coleccion de  vectores son ortonormales siempre y cuando la magnitud de cada uno sea 1

#Forma demasiada directa 
cosAngle = lambda a, b: (np.dot(a.T,b)/np.sqrt(np.dot(np.dot(a.T,a),np.dot(b.T,b))))[0][0]
print("coseno del angulo",cosAngle(vectorToMatricial1,vectorToMatricial2))

#Aqui se pasa a grados por que se encuentra en radianes y en forma matematica es cos-1(CosenoDelangulo) y el resultado ahora si sera nuestro angulo 
angle = lambda cos: np.arccos(cos)*(180/np.pi)
print("angulo",angle(cosAngle(vectorToMatricial1,vectorToMatricial2)))

#Aqui vemos otra forma de obtener el resulta 
print("Forma alternativa,Coseno del angulo",producto_punto/(longitud1*longitud2))
print("Forma altenativa angulo",angle(producto_punto/(longitud1*longitud2)))
```

Formula alterniva 
Producto punto de A y B dividido entre la multipliacion de la magnitud de a por la de b y el resultado es el coseno del angulo

```python
matriz = np.array([[1,4,5,8],[4,7,8,10],[15,4,8,2],[5,3,9,17]])
print("Matriz\n",matriz)
print("\n")
print("Dimensiones",matriz.shape)
```

```python
determinante = np.linalg.det(matriz)
print("Determinate",determinante)
```

```python
#La traza es la suma de los valores de la diagonal principal 
traza = np.trace(matriz)
print(traza)
```

```python
#El rango de una matriz es el número de filas (o columnas) linealmente independientes. 
rango = np.linalg.matrix_rank(matriz)
print("rango",rango)
```

```python
# https://www.youtube.com/watch?v=KkEuT0u1BC0&ab_channel=Integrando
# https://www.youtube.com/watch?v=_1lXryL9h9o&ab_channel=UCAMUniversidadCat%C3%B3licadeMurcia
coeff = np.poly(matriz)
print(coeff)
#Aqui tenenemos el polimonio caracteristico 
#El grado viene con consonancia con el orden de la matriz aquie vemos que tenemos un grado 4 
```

```python
#Lo que realmente pasa aqui este la ecuacion que tuvimos atras la igualamos a cero y obtemos los valores de lambda 
# esto es lo que succe  1x**4.  -33x**3.  132x**2.  133x+. -147 = 0
# aqui este ejercicio los valores de x (lambda) se le conocen como valores propios de la matriz(autovalores)
raices = np.roots(coeff)
print(raices)
```

```python
raicesAndAutovectores = np.linalg.eig(matrizA)
print(raicesAndAutovectores)
```

```python

autovector_1= np.linalg.eig(matrizA)[1][:,1]
autovector_2= np.linalg.eig(matrizA)[1][:,0]
print(autovector_1)
print(autovector_2)
#The syntax is x[row_index,column_index]
#So what that line did is sliced the array, taking all rows () but keeping the second column (:1)
```

```python
# Figure setup.
fig = plt.figure()
ax = fig.add_subplot(111)
scale = 25
origin2D = np.vstack([0,0])
plt.axis([-scale/2,scale/2,-scale/2,scale/2])
#plt.title("Vectores Original  y autovectores generados.")
plt.quiver(*origin2D,*matrizA[:,0],scale=scale, label='Vectores Originales')
plt.quiver(*origin2D,*matrizA[:,1],scale=scale)
# Figure setup.


# Eigenvectors.

plt.quiver(*origin2D, *(autovector_1*8), scale=scale,color="red", label='Autovectores_1')
plt.quiver(*origin2D, *(autovector_2*4), scale=scale,color="blue", label='Autovectores_2')


ax.set_aspect('equal', adjustable='box')
plt.legend();
```
