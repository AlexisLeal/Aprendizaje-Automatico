```python
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
linked_complete = linkage(X, 'complete')
```

```python
plt.figure(figsize=(10, 10))
dendrogram(linked_complete,
  orientation='top',
    distance_sort='descending',
    show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
rg_de_grupos= range(2,10)
val_silhouette = []
```

```python
for numero_clusters in rg_de_grupos:
 modelo = AgglomerativeClustering(
       affinity = 'euclidean',
       linkage = 'complete',
       n_clusters = numero_clusters
      )

 etiquetas_cluster = modelo.fit_predict(X)
 silhouette_avg = silhouette_score(X, etiquetas_cluster)
 val_silhouette.append(silhouette_avg)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 5))
ax.plot(rg_de_grupos, val_silhouette, marker='o')
ax.set_title("Indices silhouette")
ax.set_xlabel('Clústeres')
ax.set_ylabel('Media de los índices silhouette');
```

```python
from fcmeans import FCM
```

```python
#Se utilizan dos clusters por que es el mas optimo por su valor silhouette
fcm = FCM(n_clusters=2)
```

```python
fcm.fit(X)
```

```python
fcm_centers = fcm.centers
fcm_labels = fcm.predict(X)

# plot result
f, axes = plt.subplots(1, 2, figsize=(11,5))
axes[0].scatter(X[:,0], X[:,1])
axes[1].scatter(X[:,0], X[:,1], c=fcm_labels)
axes[1].scatter(fcm_centers[:,0], fcm_centers[:,1], marker="+", s=500, c='orange')
plt.show()
```

```python
fcm_labels
```

```python
fcm = FCM(n_clusters=3)
fcm.fit(X)
fcm_centers = fcm.centers
fcm_labels = fcm.predict(X)
f, axes = plt.subplots(1, 2, figsize=(11,5))
axes[0].scatter(X[:,0], X[:,1])
axes[1].scatter(X[:,0], X[:,1], c=fcm_labels)
axes[1].scatter(fcm_centers[:,0], fcm_centers[:,1], marker="+", s=800, c='orange')
plt.show()

fcm_labels
```
